```python
directory = "/home/grace/Projects/Dog_Vs_Cat/train"
path = "/home/grace/Projects/Dog_Vs_Cat/Train_data"
```

```python
import os
import shutil
import cv2
import matplotlib.pyplot as plt
```

```python
i = 1
j = 1
for file in os.listdir(directory):
    
    label = file.split(".")[0]
    if label == "cat":
        shutil.copy(directory + "/" + file, path + "/" + "0_" + str(i) + ".jpg")
        i = i + 1
    else:
        shutil.copy(directory + "/" + file, path + "/" + "1_" + str(j) + ".jpg")
        j = j + 1
    
```

```python
file = open("Train_label","w")
for files in os.listdir(path):
    label = files.split("_")[0]
    if label == "0":
        file.write(path + "/" + files + " 0" + "\n")
    else:
        file.write(path + "/" + files + " 1" + "\n")
file.close()
    
```
